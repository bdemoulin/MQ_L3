```python
# coding:utf-8
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Variables, in atomic units
l = 1
mu = 2
```

```python
# Plots
rho = np.linspace(0,2,1000)
eq_sym = -1 + 2*rho/mu
eq_antisym = 1 - 2*rho/mu
fonction = np.exp(-2*rho*l)
```

```python
%pylab inline

# label=u'' is for getting the accents in the legend
# r'' for having math-style display
plt.plot(rho,eq_sym,label=u'Solution symétrique')
plt.plot(rho,eq_antisym,label=u'Solution antisymétrique')
plt.plot(rho,fonction,label=r'$e^{-2 \rho l}$')

plt.xlabel(r'$\rho$',fontsize=14)
#plt.ylim([0,1.0])
plt.legend(loc='upper right',fontsize='small')
plt.grid()
plt.title(r'Possible values for $\rho$')

plt.savefig('solutions.png',dpi=300)
plt.show()
```

```python
# Plot of the wavefunctions
from scipy.optimize import brenth

# symetric_solution gives the value of rho for which the red curve crosses the blue one.
def symetric_solution(rho): 
    return np.exp(-2*rho*l) + (1 - 2*rho/mu)

def antisymetric_solution(rho):
    return np.exp(-2*rho*l) - (1 - 2*rho/mu)

# rho_s is the value of rho for the ground state
rho_s = brenth(symetric_solution,1.0,1.2)
rho_a = brenth(antisymetric_solution,0.7,0.9)

# Normalisation constants
B_s = 1 / (np.sqrt(2/rho_s * (2*l*rho_s + 1 + np.exp(2*l*rho_s))))
A_s = B_s * (1 + np.exp(2*rho_s*l))
B_a = 1 / (np.sqrt(2/rho_s * (-2*l*rho_s - 1 + np.exp(2*l*rho_s))))
A_a = B_a * (1 - np.exp(2*rho_s*l))

# Definition of the wavefunction for the three intervals
def psi_s_1(x):
    return A_s*np.exp(x)

def psi_s_2(x):
    return B_s*(np.exp(x) + np.exp(-x))

def psi_s_3(x):
    return psi_s_1(-x)

def psi_a_1(x):
    return A_a*np.exp(x)

def psi_a_2(x):
    return B_a*(np.exp(x) - np.exp(-x))

def psi_a_3(x):
    return -psi_a_1(-x)
```

```python
# Definition of the three intervals for plotting the wavefunction
x_1 = rho_s*np.linspace(-5,-l,1000)
x_2 = rho_s*np.linspace(-l,l,1000)
x_3 = rho_s*np.linspace(l,5,1000)

# Whole interval
x = np.append(x_1,x_2)
x = np.append(x,x_3)


# Functions on the whole interval
symetric = np.append(psi_s_1(x_1),psi_s_2(x_2))
symetric = np.append(symetric,psi_s_3(x_3))

antisymetric = np.append(psi_a_1(x_1),psi_a_2(x_2))
antisymetric = np.append(antisymetric,psi_a_3(x_3))

# Commands for the plot
fig, ax = plt.subplots()
ax.axhline(0, color='black')
ax.axvline(0, color='black')

# Plot the potential (delta function)
ax.axvline(x=-rho_s*l,ymin=0.50,ymax=1, color='red',linewidth=2)
ax.axvline(rho_s*l, ymin=0.50,ymax=1, color='red',linewidth=2,label='Potential')

# Plot the wavefunctions
color_s = 'blue'
color_a = 'green'
ax.set_xlabel(r'$\rho x$',fontsize=14)
ax.plot(x,symetric,color=color_s,label='Symetric wavefunction')
ax.plot(x,antisymetric,color=color_a,label='Antisymetric wavefunction')
ax.set_ylim([-1,1])
ax.grid()
ax.set_title('Plot of the wavefunctions')
ax.legend(loc='lower right',fontsize='small')

plt.savefig('wavefunction.png',dpi=300)
```

```python
# Probability
symetric_probability = symetric**2
antisymetric_probability = antisymetric**2

# Commands for the plot
fig, ax = plt.subplots()
ax.axhline(0, color='black')
ax.axvline(0, color='black')

# Plot the potential (delta function)
ax.axvline(x=-rho_s*l, color='red',linewidth=2)
ax.axvline(rho_s*l,color='red',linewidth=2,label='Potential')

# Plot the wavefunctions
color_s = 'blue'
color_a = 'green'
ax.set_xlabel(r'$\rho x$',fontsize='14')
ax.plot(x,symetric_probability,color=color_s,label='Symetric probability')
ax.plot(x,antisymetric_probability,color=color_a,label='Antisymetric probability')
ax.set_ylim([0,0.5])
ax.grid()
ax.legend(loc='upper right',fontsize='small')
ax.set_title('Plot of the probability')

plt.savefig('proba.png',dpi=300)
```

## Étude de l'énergie

```python
# Positions of the solutions
def fonction_range(rho,ll):
    return np.exp(-2*rho*ll)

plt.plot(rho,eq_sym,label=u'Solution symétrique')
plt.plot(rho,eq_antisym,label=u'Solution antisymétrique')

range_l = np.linspace(0.6,2.0,10)

for ll in range_l:

    plt.plot(rho,fonction_range(rho,ll),label=r'$e^{-2 \rho l}$')

plt.xlabel(r"$\rho$")
```

```python
# Range of lengths between the potentials
range_l = np.linspace(0.6,3.0,100)

def symetric_solution_range(rho,l): 
    return np.exp(-2*rho*l) + (1 - 2*rho/mu)

def antisymetric_solution_range(rho,l):
    return np.exp(-2*rho*l) - (1 - 2*rho/mu)

rho_s_list = []
rho_a_list = []

# Computation of rho_s for each value in range_l
for ll in range_l:
    rho_s_list.append(brenth(symetric_solution_range,1.0,2.0,args=(ll,)))
    rho_a_list.append(brenth(antisymetric_solution_range,0.1,1.0,args=(ll,)))

rho_s_table = np.array(rho_s_list)
rho_a_table = np.array(rho_a_list)
```

```python
alpha = 1
E_s = -alpha/mu * rho_s_table**2
E_a = -alpha/mu * rho_a_table**2
```

```python
plt.plot(range_l,E_s,label="Energie de la solution symetrique")
plt.plot(range_l,E_a,label="Energie de la solution antisymetrique")
plt.xlabel("$l$")
plt.ylabel("$E$")
plt.title("Energy variation with respect to l")
plt.legend(loc="upper right")
```
